```python
# Importar bibliotecas necessárias

import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException
from datetime import date
```

```python
#OBS: Essa celula só pode ser executada caso você tenha a planilha com o historico dos valores do Scraping 

# Ler planilha com histórico de cotações e cripto
cotacao_original = pd.read_excel('Scraping_financeiro.xlsx', sheet_name='Cotação')
cripto_original = pd.read_excel('Scraping_financeiro.xlsx', sheet_name='Cripto')

# Exibir dataframes
display(cotacao_original)
display(cripto_original)
```

```python
# Obtem data atual
hoje = date.today()
hoje = hoje.strftime("%d-%m-%Y")

# Configura opções do navegador
option=Options()
option.add_argument('--headless')
driver = webdriver.Firefox(options=option)

# Maximiza janela do navegador
driver.maximize_window()

# URL da página de conversão de moedas
url = 'https://www.xe.com/pt/currencyconverter/'

# Acessa URL
driver.get(url)
# Aguarda 3 segundos
time.sleep(3)

# Seleciona moeda BRL
driver.find_element('xpath', 
                       '//*[@id="midmarketFromCurrency"]/div[2]/div/input').click()
driver.find_element('xpath', 
                       '//*[@id="midmarketFromCurrency"]/div[2]/div/input').send_keys('brl')
time.sleep(1)

driver.find_element('xpath', 
                       '//*[@id="midmarketFromCurrency"]/div[2]/div/input').send_keys(Keys.ENTER)

# inverte a ordem da converção
time.sleep(1)
driver.find_element('xpath', 
                       '//*[@id="__next"]/div[4]/div[3]/section/div[1]/div[1]/div[1]/div/div/div[1]').click()
time.sleep(1)

# Obtem conteúdo HTML da tabela de cotações
element1= driver.find_element('xpath', 
                       '//*[@id="__next"]/div[4]/div[3]/section/div[1]/div[3]')
html_cotent1 = element1.get_attribute('outerHTML')
time.sleep(2)

# URL da página de criptomoedas
url = 'https://br.investing.com/crypto/'

# Acessa a URL
driver.get(url)
time.sleep(1)

# Obtem conteúdo HTML da tabela de criptomoedas
element2= driver.find_element('xpath', 
                       '//*[@id="DataTables_Table_0"]/tbody')

html_cotent2 = element2.get_attribute('outerHTML')
time.sleep(2)

# Fecha navegador
driver.quit()
```

```python
# Parsear conteúdo HTML com BeautifulSoup
soup1 = BeautifulSoup(html_cotent1,'html.parser')

# Extrai nomes e valores das moedas
name = soup1.find_all('div', class_='bqqhHY')
value= soup1.find_all('div', class_='ellipser')


z=0
# Criar listas para armazenar dados
moedas=[]
valores=[]
variacao=[]

# Itera sobre os nomes e valores das moedas
for moeda in name:
    moedas.append(moeda.get_text())

for i in list(range(0, len(value), 2)):
    valores.append(value[i].get_text())
    variacao.append(value[i+1].get_text())

# Cria o dataframe de cotações
cotacao = {
    'nome': moedas,
    'valor': valores,
    'variacao %': variacao,
    'data':[hoje]*len(variacao)
}
cotacao = pd.DataFrame(cotacao)
print('Valor das moedas em Real')
display(cotacao)
```

```python
# Parsea o conteúdo HTML com BeautifulSoup
soup2 = BeautifulSoup(html_cotent2,'html.parser')

# Extrai dados da tabela de criptomoedas
rows = soup2.find('tbody').find_all('tr')

# Cria lista para armazenar dados
cripto = []

# Itera sobre linhas da tabela
for row in rows:
    # Extrai colunas
    cols = row.find_all('td')
    
    # Cria dicionário para armazenar dados da linha
    row_data = {}
    
    # Itera sobre colunas
    for col in cols:
        # Extrai a coluna texto e classe
        text = col.get_text(strip=True)
        cls = col.get('class')[0]
        
        # Armazena dados da coluna no dicionário
        row_data[cls] = text
    
    # Adiciona dados da linha à lista
    cripto.append(row_data)

# Cria o dataframe de criptomoedas
cripto = pd.DataFrame(cripto)
cripto['data']=[hoje]*len(cripto)
cripto=cripto.drop(['flag'],axis=1)

# Renomeia as colunas do dataframe
cripto = cripto.rename(columns={
    'rank':'Ranque',
    'left': 'Simbolo',
    'price': 'Preço',
    'js-market-cap': 'Valor de mercado',
    'js-24h-volume': '24h Volume',
    'js-total-vol': 'Volume Total',
    'js-currency-change-24h': 'Mudança 24h',
    'js-currency-change-7d': 'mudança 7 dias'
})
print('Rank Das Cripto')
display(cripto)
```

```python
#OBS: Essa celula só pode ser executada caso você tenha a planilha com o historico dos valores do Scraping 

# Concatena os dados antigos com os novos
cotacao =pd.concat([cotacao,cotacao_original])
cotacao = cotacao.reset_index(drop=True)

cripto = pd.concat([cripto,cripto_original])
cripto = cripto.reset_index(drop=True)

display(cotacao)
display(cripto)
```

```python
# Escreve os dados em planilha Excel
writer = pd.ExcelWriter('Scraping_financeiro.xlsx', engine='xlsxwriter')

# Escreve os dataframes em planilhas diferentes
cotacao.to_excel(writer, sheet_name='Cotação',index=False)
cripto.to_excel(writer, sheet_name='Cripto',index=False)

# Salva planilha Excel
writer.save()
```
